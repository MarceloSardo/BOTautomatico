```python
import yfinance as yf
import pyautogui
import pyperclip

# Buscando historico de ações.

ticker = input('Digite o código da ação: ')
dados = yf.Ticker(ticker)
dados.history()
```

```python
# Definindo para 6 meses.
tabela = dados.history("6mo")
tabela
```

```python
# Selecionando coluna CLOSE
fechamento = tabela.Close
fechamento
```

```python
#Gerando gráfico de linha
fechamento.plot()
```

```python
# Gerando estatisticas
maxima = fechamento.max()
minima = fechamento.min()
atual = fechamento[-1]

print(maxima)
print(minima)
print(atual)
```

```python
# Criando e-mail 
destinatario = "marc.sardo1940@gmail.com"
assunto = "Análise diária das ações"
mensagem = f"""
Bom dia,
Segue abaixo as análises da ação {ticker} dos últimos seis meses:

Cotação máxima: R${round(maxima,2)}
Cotação mínima: R${round(minima,2)}
Cotação atual: R${round(atual,2)}

Atte,
Marcelo Lopes.
"""

print(destinatario)
print(assunto)
print(mensagem)
```
